```python
import pandas as pd

x = pd.read_csv('heart.csv')

print (x)
def leer_datos(dataset):
    data = pd.read_csv(dataset) 
    x = data.iloc[:,:-1]
    y = data.iloc[:,-1]
    return x,y
```

```python
from sklearn.model_selection import StratifiedKFold
csv_x, csv_y = leer_datos('heart.csv')
k_folds = []
# print (csv_.values)
kf = StratifiedKFold(n_splits=3)
kf
for train_index, test_index in kf.split(csv_x.values, csv_y.values):
    X_train, X_test, y_train, y_test = csv_x.values[train_index], csv_x.values[test_index], \
                                        csv_y.values[train_index], csv_y.values[test_index]
    print (X_train.shape, X_test.shape)
    k_folds.append(X_train)
    k_folds.append(y_train)
    k_folds.append(X_test)
    k_folds.append(y_test)
# def get_accuracy(model, X_train, y_train, X_test, y_test):

scores_mlp = []
print (X_train.shape)
# _y_train = np.asmatrix(y_train).T
print (X_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
index = 0
for i in k_folds:
#     print (i)
    pd.DataFrame(i).to_csv('folds' + str(index) + '.csv')
    index = index + 1
```
